```python
import pandas as pd
df = pd.read_csv('datasets/kc_house_data.csv')
```

# How many properties are available for purchase?

```python
print('There is', df.shape[0], 'properties available for purchase.')
```

# How many attributes do these properties have?

```python
print('These properties have', df.shape[1], 'attributes.')
```

# What are the attributes of these properties?


```python
print(df.columns)
```

# What is the most expensive property?

```python
print(df[['id', 'price']].sort_values('price', ascending=False))
print('The property with id 6762700020 is the most expensive.')
```

# Which property has the highest number of rooms?

```python
print(df[['id', 'bedrooms']].sort_values('bedrooms', ascending=False))
print('The property with id 2402100895 have the highest number of rooms.')
```

# What is the sum of rooms in the data set?

```python
print('The sum of rooms is equal to', sum(df['bedrooms']), '.')
```

# What is the average price of all properties?

```python
print('The average price of the properties is equal to: $', round(df['price'].mean(), 2))
```

# What is the average price of a property with 2 bathrooms?

```python
print('The average price of a property with have 2 bathrooms is: $',
      round(df.loc[df['bathrooms'] == 2, 'price'].mean(), 2))
```

# What is the minimum price between properties with 3 bedrooms?

```python
print('The minimum price between properties with 3 bedrooms is $',
      round(df.loc[df['bedrooms'] == 3, 'price'].min(), 2))
```

# How many properties have more than 300 square meters in the living room?

```python
df['m2_living'] = df['sqft_living'] * 0.092
print(df[df['m2_living'] > 300].shape[0],
      'properties have more than 300 square meters in the living room.')
```

# How many properties have more than 2 floors?

```python
print(df[df['floors'] > 2].shape[0], 'properties have more than 2 floors.')
```

# How many properties have sea view?

```python
print(df[df['waterfront'] == 1].shape[0], 'properties have sea view.')
```

# Of the properties with sea view, how many have 3 bedrooms?

```python
print(df[(df['waterfront'] == 1) & (df['bedrooms'] == 3)].shape[0],
      'of the properties with sea view have 3 bedrooms.')
```
